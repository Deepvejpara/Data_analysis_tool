```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class DataAnalyzer:
    def __init__(self, file_path=None):
        self.data = pd.DataFrame()
        self.last_plot = None
        if file_path:
            self.load_data(file_path)

    def __del__(self):
        print('Deleted Successfully!')

    def load_data(self, file_path):
        try:
            self.data = pd.read_csv(file_path)
            print('File Loaded Successfully!')
        except Exception as e:
            print(f'Error Loading Data: {e}')

    def explore_data(self):
        print('\n--- Data Exploration Menu ---')
        print('1. Show First 5 Rows')
        print('2. Show Last 5 Rows')
        print('3. Show Column Names')
        print('4. Show Data Types')
        print('5. Show Summary')

        choice = int(input('Enter your choice: '))

        if choice == 1:
            print('\n--- First 5 Rows ---')
            print(self.data.head())

        elif choice == 2:
            print('\n--- Last 5 Rows ---')
            print(self.data.tail())

        elif choice == 3:
            print('\n--- Column Names ---')
            print(self.data.columns)

        elif choice == 4:
            print('\n--- Data Types ---')
            print(self.data.dtypes)

        elif choice == 5:
            print('\n--- Basic Info ---')
            print(self.data.info())

        else:
            print('Invalid Choice!')

    def clean_data(self):
        if self.data.empty:
            print('No Data!')
            return

        print('Total Empty Data:\n', self.data.isnull().sum())

        self.data.fillna(0, inplace=True)
        print("Empty Data Replaced with 0")

        self.data.drop_duplicates(inplace=True)
        print('Duplicates Removed')

        self.data = self.data.convert_dtypes()
        print("Data types converted where applicable.")

    def static(self):
        if self.data.empty:
            print('No Data')
            return

        print('\n--- Statistical Summary ---')
        print(self.data.describe())

    def CombineData(self, other_data):
        if self.data.empty:
            print('No Data')
            return

        self.data = pd.concat([self.data, other_data], ignore_index=True)
        print('Data Combined Successfully.')

    def SplitData(self, column_name):
        if self.data.empty:
            print('No Data')
            return
        if column_name not in self.data.columns:
            print('Invalid Column')
            return

        column_groups = dict(tuple(self.data.groupby(column_name)))
        print(f"Data Split Into {len(column_groups)} groups by column '{column_name}'")
        return column_groups

    def SortFilter(self):
        if self.data.empty:
            print('No Data')
            return

        col = input('\nEnter Column Name: ')
        if col not in self.data.columns:
            print('Column not Found')
            return

        val = input('\nEnter Value to Filter: ')
        sorted_data = self.data[self.data[col].astype(str) == val]

        print('\n--- Filtered Data ---')
        print(sorted_data.head())

    def PivotTable(self):
        if self.data.empty:
            print("No Data")
            return

        index = input("Enter Column for Index: ")
        values = input("Enter Column for Values: ")

        if index not in self.data.columns or values not in self.data.columns:
            print("Columns not Found")
            return

        pivot = pd.pivot_table(self.data, index=index, values=values, aggfunc="sum")
        print('\n--- Pivot Table ---')
        print(pivot)

    def vizualize_data(self):
        if self.data.empty:
            print('No Data')
            return

        print('\n--- Choose Preferred Visualization ---')
        print('1. Line Plot')
        print('2. Bar Plot')
        print('3. Scatter Plot')
        print('4. Histogram')
        print('5. Box Plot')

        choice = int(input('Enter Your Choice: '))
        plt.figure(figsize=(10, 7))

        if choice in [1, 2, 3, 5]:
            column_x = input('Enter X Column: ')
            column_y = input('Enter Y Column: ')

            if column_x not in self.data.columns or column_y not in self.data.columns:
                print('Invalid Columns')
                return

            if choice == 1:
                self.last_plot = sns.lineplot(data=self.data, x=column_x, y=column_y)
            elif choice == 2:
                self.last_plot = sns.barplot(data=self.data, x=column_x, y=column_y)
            elif choice == 3:
                self.last_plot = sns.scatterplot(data=self.data, x=column_x, y=column_y)
            elif choice == 5:
                self.last_plot = sns.boxplot(data=self.data, x=column_x, y=column_y)

        elif choice == 4:
            x = input("Enter column for Histogram: ")
            if x not in self.data.columns:
                print("Invalid column.")
                return
            self.last_plot = self.data[x].hist()

        else:
            print('Invalid Choice')
            return

        plt.show()
      
# -- Main Menu --

def main():
    tool = DataAnalyzer()
    while True:
        print('\n--- Data Analysis Tools ---')
        print('1. Load Dataset')
        print('2. Explore Data')
        print('3. Handle Missing Data')
        print('4. Data Describtion')
        print('5. Split Data')
        print('6. Sort Data')
        print('7. Create Pivot Table')
        print('8. Data Visulize')
        print('9. Exit')

        choice = int(input('\nEnter Your Choice:'))

        if choice == 1:
            path = input('Enter Dataset Path:')
            tool.load_data(path)

        elif choice == 2:
            tool.explore_data()

        elif choice == 3:
            tool.clean_data()

        elif choice == 4:
            tool.static()

        elif choice == 5:
            column = input('Enter Column name:')
            if column:
                tool.CombineData(column)

        elif choice == 6:
            tool.SortFilter()

        elif choice == 7:
            tool.PivotTable()

        elif choice == 8:
            tool.vizualize_data()

        elif choice == 9:
            break

        else:
            print('Invalid Choice')

if __name__ == "__main__":
    main()
```
